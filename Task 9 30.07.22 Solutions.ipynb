# Q9)   Average  weekly response time for each agent.

```python
def get_seconds(time_str):                                         # Fuction to convert time in int dtype
    hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
```

```python
avg_response_sec = []                                              # Create one new column which include avg_respose_sec
for t in df['Average Response Time']:                              # means convert the average respose time into seconds
    avg_response_sec.append(get_seconds(t))
avg_response_sec
df['avg_response_sec'] = avg_response_sec 
```

```python
df.head()
```

```python
df.groupby(['Agent Name','week'])['avg_response_sec'].mean()
```

# Q10)   Average  weekly resolution time for each agent.

```python
avg_resolution_sec = []
for t in df['Average Resolution Time']:
    avg_resolution_sec.append(get_seconds(t))
avg_resolution_sec
df['avg_resolution_sec'] = avg_resolution_sec
df.head()
```
